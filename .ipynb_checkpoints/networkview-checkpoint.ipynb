```python
import json
import ipywidgets as widgets
from IPython.display import display, HTML

# Load the Neo4j results (replace with your file path or use json.loads() if you have the data in a variable)
results = [
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 9,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1964,
            "high": 0
          },
          "name": "Keanu Reeves"
        },
        "elementId": "9"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 221,
          "high": 0
        },
        "start": {
          "low": 9,
          "high": 0
        },
        "end": {
          "low": 162,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Julian Mercer"
          ]
        },
        "elementId": "221",
        "startNodeElementId": "9",
        "endNodeElementId": "162"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 162,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "title": "Something's Gotta Give",
          "released": {
            "low": 2003,
            "high": 0
          }
        },
        "elementId": "162"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 9,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1964,
            "high": 0
          },
          "name": "Keanu Reeves"
        },
        "elementId": "9"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 114,
          "high": 0
        },
        "start": {
          "low": 9,
          "high": 0
        },
        "end": {
          "low": 95,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Shane Falco"
          ]
        },
        "elementId": "114",
        "startNodeElementId": "9",
        "endNodeElementId": "95"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 95,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "Pain heals, Chicks dig scars... Glory lasts forever",
          "title": "The Replacements",
          "released": {
            "low": 2000,
            "high": 0
          }
        },
        "elementId": "95"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 9,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1964,
            "high": 0
          },
          "name": "Keanu Reeves"
        },
        "elementId": "9"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 132,
          "high": 0
        },
        "start": {
          "low": 9,
          "high": 0
        },
        "end": {
          "low": 108,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Johnny Mnemonic"
          ]
        },
        "elementId": "132",
        "startNodeElementId": "9",
        "endNodeElementId": "108"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 108,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "The hottest data on earth. In the coolest head in town",
          "title": "Johnny Mnemonic",
          "released": {
            "low": 1995,
            "high": 0
          }
        },
        "elementId": "108"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 9,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1964,
            "high": 0
          },
          "name": "Keanu Reeves"
        },
        "elementId": "9"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 22,
          "high": 0
        },
        "start": {
          "low": 9,
          "high": 0
        },
        "end": {
          "low": 19,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Kevin Lomax"
          ]
        },
        "elementId": "22",
        "startNodeElementId": "9",
        "endNodeElementId": "19"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 19,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "Evil has its winning ways",
          "title": "The Devil's Advocate",
          "released": {
            "low": 1997,
            "high": 0
          }
        },
        "elementId": "19"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 9,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1964,
            "high": 0
          },
          "name": "Keanu Reeves"
        },
        "elementId": "9"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 15,
          "high": 0
        },
        "start": {
          "low": 9,
          "high": 0
        },
        "end": {
          "low": 18,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Neo"
          ]
        },
        "elementId": "15",
        "startNodeElementId": "9",
        "endNodeElementId": "18"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 18,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "Everything that has a beginning has an end",
          "title": "The Matrix Revolutions",
          "released": {
            "low": 2003,
            "high": 0
          }
        },
        "elementId": "18"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 9,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1964,
            "high": 0
          },
          "name": "Keanu Reeves"
        },
        "elementId": "9"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 8,
          "high": 0
        },
        "start": {
          "low": 9,
          "high": 0
        },
        "end": {
          "low": 17,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Neo"
          ]
        },
        "elementId": "8",
        "startNodeElementId": "9",
        "endNodeElementId": "17"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 17,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "Free your mind",
          "title": "The Matrix Reloaded",
          "released": {
            "low": 2003,
            "high": 0
          }
        },
        "elementId": "17"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 9,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1964,
            "high": 0
          },
          "name": "Keanu Reeves"
        },
        "elementId": "9"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 0,
          "high": 0
        },
        "start": {
          "low": 9,
          "high": 0
        },
        "end": {
          "low": 8,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Neo"
          ]
        },
        "elementId": "0",
        "startNodeElementId": "9",
        "endNodeElementId": "8"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 8,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "Welcome to the Real World",
          "title": "The Matrix",
          "released": {
            "low": 1999,
            "high": 0
          }
        },
        "elementId": "8"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 10,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1967,
            "high": 0
          },
          "name": "Carrie-Anne Moss"
        },
        "elementId": "10"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 16,
          "high": 0
        },
        "start": {
          "low": 10,
          "high": 0
        },
        "end": {
          "low": 18,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Trinity"
          ]
        },
        "elementId": "16",
        "startNodeElementId": "10",
        "endNodeElementId": "18"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 18,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "Everything that has a beginning has an end",
          "title": "The Matrix Revolutions",
          "released": {
            "low": 2003,
            "high": 0
          }
        },
        "elementId": "18"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 10,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1967,
            "high": 0
          },
          "name": "Carrie-Anne Moss"
        },
        "elementId": "10"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 9,
          "high": 0
        },
        "start": {
          "low": 10,
          "high": 0
        },
        "end": {
          "low": 17,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Trinity"
          ]
        },
        "elementId": "9",
        "startNodeElementId": "10",
        "endNodeElementId": "17"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 17,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "Free your mind",
          "title": "The Matrix Reloaded",
          "released": {
            "low": 2003,
            "high": 0
          }
        },
        "elementId": "17"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  },
  {
    "keys": [
      "anchor",
      "anchorLabels",
      "rel",
      "relType",
      "neighbor",
      "neighborLabels"
    ],
    "length": 6,
    "_fields": [
      {
        "identity": {
          "low": 10,
          "high": 0
        },
        "labels": [
          "Person"
        ],
        "properties": {
          "born": {
            "low": 1967,
            "high": 0
          },
          "name": "Carrie-Anne Moss"
        },
        "elementId": "10"
      },
      [
        "Person"
      ],
      {
        "identity": {
          "low": 1,
          "high": 0
        },
        "start": {
          "low": 10,
          "high": 0
        },
        "end": {
          "low": 8,
          "high": 0
        },
        "type": "ACTED_IN",
        "properties": {
          "roles": [
            "Trinity"
          ]
        },
        "elementId": "1",
        "startNodeElementId": "10",
        "endNodeElementId": "8"
      },
      "ACTED_IN",
      {
        "identity": {
          "low": 8,
          "high": 0
        },
        "labels": [
          "Movie"
        ],
        "properties": {
          "tagline": "Welcome to the Real World",
          "title": "The Matrix",
          "released": {
            "low": 1999,
            "high": 0
          }
        },
        "elementId": "8"
      },
      [
        "Movie"
      ]
    ],
    "_fieldLookup": {
      "anchor": 0,
      "anchorLabels": 1,
      "rel": 2,
      "relType": 3,
      "neighbor": 4,
      "neighborLabels": 5
    }
  }
]

# Process the results to extract nodes and links for the network visualization
nodes = {}
links = []

for result in results:
    # Process anchor node
    anchor = result['_fields'][0]
    anchor_id = anchor['elementId']
    anchor_labels = result['_fields'][1]
    
    if anchor_id not in nodes:
        # Use name property as label if available, otherwise use ID
        label = anchor['properties'].get('name', f"ID: {anchor_id}")
        nodes[anchor_id] = {
            'id': anchor_id,
            'label': label,
            'labels': anchor_labels,
            'properties': anchor['properties'],
            'type': 'anchor'  # To distinguish anchor nodes
        }
    
    # Process neighbor node
    neighbor = result['_fields'][4]
    neighbor_id = neighbor['elementId']
    neighbor_labels = result['_fields'][5]
    
    if neighbor_id not in nodes:
        # Use title property as label for movies, name for persons, or ID
        if 'title' in neighbor['properties']:
            label = neighbor['properties']['title']
        elif 'name' in neighbor['properties']:
            label = neighbor['properties']['name']
        else:
            label = f"ID: {neighbor_id}"
            
        nodes[neighbor_id] = {
            'id': neighbor_id,
            'label': label,
            'labels': neighbor_labels,
            'properties': neighbor['properties'],
            'type': 'neighbor'
        }
    
    # Process relationship
    rel = result['_fields'][2]
    rel_type = result['_fields'][3]
    
    links.append({
        'source': anchor_id,
        'target': neighbor_id,
        'type': rel_type,
        'properties': rel['properties']
    })

# Convert to lists for the visualization
nodes_list = list(nodes.values())
    
# Create the widget HTML and JavaScript for the interactive network
html_content = """
<div style="display: flex; width: 100%;">
    <div style="flex: 3;">
        <canvas id="networkCanvas" width="600" height="400" style="border: 1px solid #ccc;"></canvas>
        <div id="controls" style="margin-top: 10px;">
            <button id="zoomIn">Zoom In</button>
            <button id="zoomOut">Zoom Out</button>
            <button id="resetView">Reset View</button>
        </div>
    </div>
    <div style="flex: 2; padding-left: 20px;">
        <div id="nodeInfo">
            <h3>Node Information</h3>
            <p>Click on a node to see its details</p>
            <div id="nodeProperties"></div>
        </div>
    </div>
</div>

<script>
(function() {
    // Network visualization data
    const nodes = """ + json.dumps(nodes_list) + """;
    const links = """ + json.dumps(links) + """;
    
    // Canvas setup
    const canvas = document.getElementById('networkCanvas');
    const ctx = canvas.getContext('2d');
    const width = canvas.width;
    const height = canvas.height;
    
    // Visual properties
    const nodeRadius = 12;
    const anchorColor = '#ff6347';  // Tomato red for anchor nodes
    const neighborColor = '#4682b4'; // Steel blue for neighbor nodes
    let scale = 1;
    let offsetX = width / 2;
    let offsetY = height / 2;
    let selectedNode = null;
    
    // Force-directed layout variables
    let nodePositions = {};
    let nodeVelocities = {};
    
    // Initialize node positions and velocities
    function initializePositions() {
        nodes.forEach(node => {
            // Assign random initial positions
            nodePositions[node.id] = {
                x: Math.random() * width,
                y: Math.random() * height
            };
            
            // Initialize velocities to zero
            nodeVelocities[node.id] = { x: 0, y: 0 };
        });
    }
    
    // Initialize
    initializePositions();
    
    // Force-directed layout simulation
    function runSimulation() {
        const repulsionForce = 200;  // Repulsion force between nodes
        const linkDistance = 80;     // Ideal link distance
        const linkStrength = 0.08;   // Link force strength
        const damping = 0.8;        // Velocity damping factor
        const centerForce = 0.01;    // Force pulling nodes to center
        
        // Apply forces to each node
        nodes.forEach(node => {
            let fx = 0, fy = 0;
            
            // Repulsion between nodes
            nodes.forEach(otherNode => {
                if (node.id !== otherNode.id) {
                    const dx = nodePositions[node.id].x - nodePositions[otherNode.id].x;
                    const dy = nodePositions[node.id].y - nodePositions[otherNode.id].y;
                    const distance = Math.sqrt(dx * dx + dy * dy) || 1;
                    
                    // Repulsion force inversely proportional to distance
                    const force = repulsionForce / (distance * distance);
                    fx += dx * force / distance;
                    fy += dy * force / distance;
                }
            });
            
            // Link forces
            links.forEach(link => {
                if (link.source === node.id || link.target === node.id) {
                    const targetId = link.source === node.id ? link.target : link.source;
                    const dx = nodePositions[node.id].x - nodePositions[targetId].x;
                    const dy = nodePositions[node.id].y - nodePositions[targetId].y;
                    const distance = Math.sqrt(dx * dx + dy * dy) || 1;
                    
                    // Attraction/repulsion to maintain link distance
                    const force = linkStrength * (linkDistance - distance);
                    fx -= dx * force / distance;
                    fy -= dy * force / distance;
                }
            });
            
            // Center force to keep nodes from drifting too far
            fx -= centerForce * (nodePositions[node.id].x - width/2);
            fy -= centerForce * (nodePositions[node.id].y - height/2);
            
            // Update velocity with damping
            nodeVelocities[node.id].x = (nodeVelocities[node.id].x + fx) * damping;
            nodeVelocities[node.id].y = (nodeVelocities[node.id].y + fy) * damping;
            
            // Update position
            nodePositions[node.id].x += nodeVelocities[node.id].x;
            nodePositions[node.id].y += nodeVelocities[node.id].y;
            
            // Boundary constraints
            nodePositions[node.id].x = Math.max(nodeRadius, Math.min(width - nodeRadius, nodePositions[node.id].x));
            nodePositions[node.id].y = Math.max(nodeRadius, Math.min(height - nodeRadius, nodePositions[node.id].y));
        });
    }
    
    // Run a few iterations to settle the layout
    for (let i = 0; i < 100; i++) {
        runSimulation();
    }
    
    // Draw the network
    function drawNetwork() {
        ctx.clearRect(0, 0, width, height);
        ctx.save();
        
        // Apply zoom and pan
        ctx.translate(offsetX, offsetY);
        ctx.scale(scale, scale);
        ctx.translate(-width/2, -height/2);
        
        // Draw links first (so they're behind nodes)
        links.forEach(link => {
            const sourcePos = nodePositions[link.source];
            const targetPos = nodePositions[link.target];
            
            ctx.beginPath();
            ctx.moveTo(sourcePos.x, sourcePos.y);
            ctx.lineTo(targetPos.x, targetPos.y);
            ctx.strokeStyle = '#999';
            ctx.lineWidth = 1;
            ctx.stroke();
            
            // Draw relationship type in the middle of the link
            const midX = (sourcePos.x + targetPos.x) / 2;
            const midY = (sourcePos.y + targetPos.y) / 2;
            ctx.fillStyle = '#666';
            ctx.font = '8px Arial';
            ctx.fillText(link.type, midX, midY);
        });
        
        // Draw nodes
        nodes.forEach(node => {
            const pos = nodePositions[node.id];
            
            // Draw circle
            ctx.beginPath();
            ctx.arc(pos.x, pos.y, nodeRadius, 0, Math.PI * 2);
            
            // Different color for anchor vs neighbor nodes
            ctx.fillStyle = node.type === 'anchor' ? anchorColor : neighborColor;
            
            // Highlight selected node
            if (selectedNode && selectedNode.id === node.id) {
                ctx.lineWidth = 3;
                ctx.strokeStyle = '#ffd700'; // gold highlight
                ctx.stroke();
            }
            
            ctx.fill();
            
            // Draw node label
            ctx.fillStyle = '#000';
            ctx.font = '10px Arial';
            ctx.textAlign = 'center';
            ctx.fillText(node.label, pos.x, pos.y + nodeRadius + 12);
            
            // Draw node type/label above the node
            if (node.labels && node.labels.length > 0) {
                ctx.font = '8px Arial';
                ctx.fillText(node.labels.join(', '), pos.x, pos.y - nodeRadius - 5);
            }
        });
        
        ctx.restore();
    }
    
    // Initial draw
    drawNetwork();
    
    // Handle canvas click to select nodes
    canvas.addEventListener('click', (event) => {
        const rect = canvas.getBoundingClientRect();
        const clickX = event.clientX - rect.left;
        const clickY = event.clientY - rect.top;
        
        // Convert click coordinates to canvas space, accounting for zoom and pan
        const canvasX = (clickX - offsetX) / scale + width/2;
        const canvasY = (clickY - offsetY) / scale + height/2;
        
        // Find clicked node
        selectedNode = null;
        
        for (const node of nodes) {
            const pos = nodePositions[node.id];
            const dx = pos.x - canvasX;
            const dy = pos.y - canvasY;
            const distance = Math.sqrt(dx*dx + dy*dy);
            
            if (distance <= nodeRadius) {
                selectedNode = node;
                break;
            }
        }
        
        // Update display and redraw
        updateNodeInfo();
        drawNetwork();
    });
    
    // Update the HTML table with node properties
    function updateNodeInfo() {
        const propertiesDiv = document.getElementById('nodeProperties');
        
        if (!selectedNode) {
            propertiesDiv.innerHTML = '<p>Click on a node to see its details</p>';
            return;
        }
        
        let html = `
            <h4>${selectedNode.label} <small>(${selectedNode.labels.join(', ')})</small></h4>
            <table style="border-collapse: collapse; width: 100%;">
                <tr>
                    <th style="border: 1px solid #ddd; padding: 8px; text-align: left; background-color: #f2f2f2;">Property</th>
                    <th style="border: 1px solid #ddd; padding: 8px; text-align: left; background-color: #f2f2f2;">Value</th>
                </tr>
        `;
        
        // Add each property to the table
        for (const [key, value] of Object.entries(selectedNode.properties)) {
            let displayValue = value;
            
            // Handle different value types
            if (typeof value === 'object' && value !== null) {
                if ('low' in value && 'high' in value) {
                    // Handle Neo4j integer objects
                    displayValue = value.low;
                } else {
                    // Handle other objects
                    displayValue = JSON.stringify(value);
                }
            } else if (Array.isArray(value)) {
                // Handle arrays
                displayValue = value.join(', ');
            }
            
            html += `
                <tr>
                    <td style="border: 1px solid #ddd; padding: 8px;">${key}</td>
                    <td style="border: 1px solid #ddd; padding: 8px;">${displayValue}</td>
                </tr>
            `;
        }
        
        html += '</table>';
        
        // Add related links information
        html += `<h4>Relationships:</h4><ul>`;
        
        // Find links connected to this node
        const nodeLinks = links.filter(link => 
            link.source === selectedNode.id || link.target === selectedNode.id
        );
        
        if (nodeLinks.length === 0) {
            html += '<li>No relationships</li>';
        } else {
            nodeLinks.forEach(link => {
                const isSource = link.source === selectedNode.id;
                const otherNodeId = isSource ? link.target : link.source;
                const otherNode = nodes.find(n => n.id === otherNodeId);
                const direction = isSource ? 'to' : 'from';
                
                html += `<li><b>${isSource ? '→' : '←'} ${link.type}</b> ${direction} ${otherNode.label}</li>`;
            });
        }
        
        html += '</ul>';
        propertiesDiv.innerHTML = html;
    }
    
    // Zoom and pan controls
    document.getElementById('zoomIn').addEventListener('click', () => {
        scale *= 1.2;
        drawNetwork();
    });
    
    document.getElementById('zoomOut').addEventListener('click', () => {
        scale /= 1.2;
        drawNetwork();
    });
    
    document.getElementById('resetView').addEventListener('click', () => {
        scale = 1;
        offsetX = width / 2;
        offsetY = height / 2;
        drawNetwork();
    });
    
    // Mouse drag to pan
    let isDragging = false;
    let lastX, lastY;
    
    canvas.addEventListener('mousedown', (event) => {
        isDragging = true;
        lastX = event.clientX;
        lastY = event.clientY;
    });
    
    window.addEventListener('mousemove', (event) => {
        if (isDragging) {
            const dx = event.clientX - lastX;
            const dy = event.clientY - lastY;
            
            offsetX += dx;
            offsetY += dy;
            
            lastX = event.clientX;
            lastY = event.clientY;
            
            drawNetwork();
        }
    });
    
    window.addEventListener('mouseup', () => {
        isDragging = false;
    });
    
    // Mouse wheel to zoom
    canvas.addEventListener('wheel', (event) => {
        event.preventDefault();
        
        // Determine zoom direction
        const zoomFactor = event.deltaY < 0 ? 1.1 : 0.9;
        
        // Update scale
        scale *= zoomFactor;
        
        drawNetwork();
    });
})();
</script>
"""

# Create and display the widget
widget = widgets.HTML(html_content)
display(widget)
```
